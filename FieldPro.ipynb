```python
import pandas as pd
import datetime as dt
import matplotlib.pyplot as plt
import numpy as np
```

```python
# carrega o primeiro conjunto de dados, ajustando a data para o padrão UTC
stn = pd.read_csv('Estacao_Convencional.csv')
stn['datetime'] = pd.to_datetime(stn['data'] + ' ' + stn['Hora (Brasília)'])
stn = stn.drop(['data', 'Hora (Brasília)'], axis=1)
stn.set_index('datetime',inplace=True)
stn.index = stn.index + dt.timedelta(hours=3)
stn.index = stn.index.tz_localize('UTC')
```

```python
# carrega o conjunto de dados do sensor, aproveita a deixa e padroniza a coluna de data 
sensor = pd.read_csv('Sensor_FieldPRO.csv')
sensor.rename(columns={"Datetime – utc": "datetime"},inplace=True)
sensor.index = pd.to_datetime(sensor['datetime'])
sensor = sensor.drop(['datetime'],axis=1)
```

```python
# concatena os dois dataframes, compartilhando o mesmo index que é o datetime natz UTC, remove os nan
df = pd.concat([stn,sensor],axis=1)
df.dropna(inplace=True)
```

```python
# como uma métrica simples e de facil entendimento... vamos ver a matrix de correlação em um heatmap
def show_heatmap(data):
	import matplotlib.pyplot as plt
	plt.matshow(data.corr(),vmin=0,vmax=1, cmap='Spectral')
	plt.xticks(range(data.shape[1]), data.columns, fontsize=7, rotation=45)
	plt.gca().xaxis.tick_bottom()
	plt.yticks(range(data.shape[1]), data.columns, fontsize=7, rotation=45)
	for (i, j), z in np.ndenumerate(data.corr()):
		plt.text(j, i, '{:0.2f}'.format(z), ha='center', va='center')
	cb = plt.colorbar()
	cb.ax.tick_params(labelsize=10)
	plt.title(f'Correlation Heatmap', fontsize=14)
	plt.show()

show_heatmap(df)
```

```python
# epa, pera lá... tem bastante coisa interesante aí...
# existe correlaçao forte entre umidade e temperatura do sensor (0.97) [não sei o que fazer com isso]
# correlação muito boa entre a variação de carga e a chuva (0.64) 
# parece que a variação de carga tem correlação com a temperautra também (0.59) [vamos ver]
fig, ax1 = plt.subplots()
df['piezo_temperature'].plot(kind='line', color='blue', ax=ax1)
ax2 = ax1.twinx()
df['change'].plot(kind='line', color='red', ax=ax2)
ax1.yaxis.label.set_color('blue')
ax2.yaxis.label.set_color('red')
ax1.set_ylabel('piezo_temperature')
ax2.set_ylabel('change')
plt.show()
```

```python
# vamo tentar criar uma relação entre a variação de carga e as temperaturas... 
moda =  df.groupby(by='piezo_temperature')['change'].agg(pd.Series.mode)
moda
```

```python
# pegando só  primeiro elemento do grupo moda
def uniqueofgroup(element):
	if isinstance(element,np.ndarray):
		return element[0]
	else:
		return element
    
moda = moda.apply(lambda x: uniqueofgroup(x))
moda
```

```python
# jeitão de coisa linear... 
moda.plot()
```

```python
# evocando a regressão por mínimos quadrados lá do laboratório de Física I, II, III, IV...
def linear_regression(x, y):
    N = len(x)
    sum_x = np.sum(x)
    sum_y = np.sum(y)
    sum_x_squared = np.sum(np.square(x))
    sum_xy = np.sum(x * y)
    # fórmulas para calcular os coeficientes
    m = (N * sum_xy - sum_x * sum_y) / (N * sum_x_squared - sum_x ** 2)
    b = (sum_y * sum_x_squared - sum_x * sum_xy) / (N * sum_x_squared - sum_x ** 2)
    return b, m

b, m = linear_regression(moda.index,moda.values)
```

```python
# verificando 
plt.plot(moda.index*m + b)
plt.plot(moda.values)
```

```python
# calcula o valor tórico baseado no modelo de regressão linear...
df['linearmodel'] = df.apply(lambda r: (r['change'] - r['piezo_temperature']*m+b),axis=1)
```

```python
fig, ax1 = plt.subplots()
df['linearmodel'].plot(kind='line', color='blue', ax=ax1)
ax2 = ax1.twinx()
df['chuva'].plot(kind='line', color='red', ax=ax2)
ax1.yaxis.label.set_color('blue')
ax2.yaxis.label.set_color('red')
ax1.set_ylabel('linearmodel')
ax2.set_ylabel('chuva')
plt.show()
```

```python
# legal... mas acho que, para regressão, estamos interessados somente em epsódios de chuva!
df['linearmodel'] = df['linearmodel'].apply(lambda r: r if r >=0 else 0)
```

 aqui formulo algumas hipoteses que podem estar relacionada à erros induzidos no experimento...

0) A precipitação é uma variável discreta, neste caso, com intervalos de no mínimo 0.2 mm (limitado à resolução amostral do pluviômetro). E a relação que queremos encontrar é uma relação da variação de carga no piezo com a quantidade de chuva... Portanto, avalio este problema como uma questão de "classificação" diferente de uma regressão em escala de tempo, uma vez que a precitação não tem uma boa relação em escala de tempo (no curto prazo) - ao contrário do que seria observado em uma regressão para temperautra (que não é o caso aqui)...

1) essa curva de variação de carga x chuva tem cara de ser linear... mas a amostra é pequena e irregular para afirmar isso...

2) a amostra de chuva em distribuição de mm é relativamente pequena (apenas 26 classes)

3) A possível maior fonte de incerteza: O sensor e o pluviômetro de referencia podem estar muito distantes espacialmente (até mesmo questão de metros)... considerando a narureza da chuva como uma variável irregular e mal distribuída... Os registros na variação de carga podem ter sido SIM registros de chuvas reais... mas, não tem com o que a gente comparar... Como no exemplo dos 4.6 mm no pluviômetro de referência e os 501.405405 registrados no sensor... Neste caso, possivelmente, a chuva foi real e mais forte no situ do sensor do que a chuva que aconteceu no pluviômetro...

4) Registros de 0.2 mm também podem ser considerados como uma fonte de incerteza... no geral, este valor é equivalente à resolução amostral dos pluviômetros... Além disso, 0.2 mm pode ocorrer em situações de não chuva como por exemplo o registro por condensação do orvalho nas paredes do funil de coleta do pluviômetro...

```python
# postulado isso, vamos tentar criar um dataframe com os valores de refenrecia mais coerentes... 
# usando os mínimos e máximos das classes, o desvio padrão e a primeira e segunda diferencial...
df2 = df.groupby(by='chuva')['linearmodel'].mean()
df2 = df2.reset_index()
df2['min'] = df.groupby(by='chuva')['linearmodel'].min().values
df2['max'] = df.groupby(by='chuva')['linearmodel'].max().values
df2['std'] = df.groupby(by='chuva')['linearmodel'].std().values
df2['diff'] = df2['linearmodel'] - df2['linearmodel'].shift(1) # primeira derivada
df2['diff2'] = df2['diff'] - df2['diff'].shift(1) #segunda derivada
df2
```

```python
# criando um x e y das categorias de chuva...
x = entrada['linearmodel'].values.reshape(-1, 1)
y = entrada['chuva'].index

pChange = df['linearmodel'].values
chuvaReal = df['chuva'].values
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.neural_network import MLPRegressor
```

```python
# 1. Regressão Linear
model_linear = LinearRegression()
model_linear.fit(x, y)
predictions_linear = model_linear.predict(pChange.reshape(-1,1))
mse_linear = mean_squared_error(chuvaReal.reshape(-1,1), predictions_linear)
```

```python
# 2. Regressão Polinomial
poly_features = PolynomialFeatures(degree=2)
X_poly = poly_features.fit_transform(x)
model_poly = LinearRegression()
model_poly.fit(X_poly, y)
predictions_poly = model_poly.predict(poly_features.transform(pChange.reshape(-1,1)))
mse_poly = mean_squared_error(chuvaReal.reshape(-1,1), predictions_poly)
```

```python
# 3. Árvores de Decisão
model_tree = DecisionTreeRegressor()
model_tree.fit(x, y)
predictions_tree = [ model_tree.predict([[r]])[0] for r in pChange ]
mse_tree = mean_squared_error(chuvaReal.reshape(-1,1), predictions_tree)
```

```python
# 4. Florestas Aleatórias
model_forest = RandomForestRegressor(n_estimators=100, random_state=42)
model_forest.fit(x, y)
predictions_forest = [ model_forest.predict([[r]])[0] for r in pChange ]
mse_forest = mean_squared_error(chuvaReal.reshape(-1,1), predictions_forest)
```

```python
# 5. Redes Neurais
model_nn = MLPRegressor(hidden_layer_sizes=(10,10), max_iter=5000, random_state=42)
model_nn.fit(x, y)
predictions_nn = model_nn.predict(pChange.reshape(-1,1))
mse_nn = mean_squared_error(chuvaReal.reshape(-1,1), predictions_nn)
```

```python
print(f'MSE da Regressão Linear: {mse_linear}')
print(f'MSE da Regressão Polinomial: {mse_poly}')
print(f'MSE das Árvores de Decisão: {mse_tree}')
print(f'MSE das Florestas Aleatórias: {mse_forest}')
print(f'MSE das Redes Neurais: {mse_nn}')
```
